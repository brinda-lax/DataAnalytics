#Importing libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

#Read file

```python
Aids=pd.read_csv('/content/sample_data/hiv-aids-1984-to-2022-eng.csv')
Aids
```

```python
Aids.info()
```

#describe

```python
Aids_describe=Aids.describe()
Aids_describe
```

#Heatmap on all correlated values

```python
plt.figure(figsize=(15,15))
sns.heatmap(Aids_describe.corr(),annot=True,fmt='.2f')
plt.show()
```

```python
Aids.notna()
Aids.fillna(0,inplace=True)
```

#seggregating fewer data

```python
Aids_subset_male=Aids[['Year','Sex/Male','Route of Transmission/Heterosexual','Route of Transmission/Homosexual','Route of Transmission/Bisexual','Total']]
Aids_subset_male
```

#Separating year and disease

```python
Aids_subset=pd.DataFrame(Aids_subset_male)
Aids_subset[['Year','Disease']]=Aids_subset_male['Year'].str.split('/',expand=True)
Aids_subset
```

```python
group_male=Aids_subset.groupby('Disease').size().sort_values(ascending=False)
#group_male
male_pie=plt.pie(group_male,labels=group_male.index,autopct='%1.1f%%')
plt.title('Overall Male affexted from 1984-tilldate')
plt.show()
```

#Now Analysing for Female

```python
Aids_female=Aids[['Year','Sex/Female','Route of Transmission/Heterosexual','Route of Transmission/Homosexual','Route of Transmission/Bisexual','Total']]
Aids_female

```

```python
Aids_subsetf=pd.DataFrame(Aids_female)
Aids_subsetf[['Year','Disease']]=Aids_female['Year'].str.split('/',expand=True)
Aids_subsetf
```

```python
Aids_female_df=Aids_subsetf.groupby('Disease').size().sort_values(ascending=False)
Aids_female_group=Aids_female_df.reset_index(name='Count')
Aids_female_group
plt.pie(Aids_female_group['Count'],labels=Aids_female_group['Disease'],autopct='%1.1f%%')
plt.title('Overall Female affexted from 1984-tilldate')
plt.show()
```

```python
Aids_disease=pd.DataFrame(Aids)
Aids_disease[['Year','Disease']]=Aids_disease['Year'].str.split('/',expand=True)
Aids_disease.reset_index(inplace=False)
Aids_disease
#print(Aids_disease.columns)
```

```python
Aids_chinese = Aids_disease[['Year','Disease', 'Ethnicity/Chinese ']]
Aids_chinese
```

```python
Cummulative=Aids_chinese.iloc[78:80]
Cummulative
```

```python
Aids_chinese.drop([78,79],inplace=True)
Aids_chinese
```

```python
Aids_chinese_sorted=Aids_chinese.sort_values(by='Ethnicity/Chinese ',ascending=False)
chinese_aids_top=Aids_chinese_sorted.head(10)
sns.barplot(x='Year',y='Ethnicity/Chinese ',hue='Year',data=chinese_aids_top)
plt.title('')
plt.show()
```

#Non chinese

```python
Aids_nonchinese = Aids_disease[['Year','Disease', 'Ethnicity/Non-Chinese']]
nonchineseCummulative=Aids_nonchinese.iloc[78:80]
#nonchineseCummulative
Aids_nonchinese.drop([78,79],inplace=True)
Aids_nonchinese
```

```python
Aids_nonchinese_sorted=Aids_nonchinese.sort_values(by='Ethnicity/Non-Chinese',ascending=False)
nonchinese_aids_top=Aids_nonchinese_sorted.head(10)
sns.barplot(x='Year',y='Ethnicity/Non-Chinese',hue='Year',data=nonchinese_aids_top)
plt.title('Disease in Nonchinese top 10 years')
plt.show()
```

#Unknown ethnicity

```python
Aids_unknown = Aids_disease[['Year','Disease', 'Ethnicity/Unknown']]
#nonchineseCummulative=Aids_nonchinese.iloc[78:80]
#nonchineseCummulative
Aids_unknown.drop([78,79],inplace=True)
#Aids_unknown
Aids_unknown_sorted=Aids_unknown.sort_values(by='Ethnicity/Unknown',ascending=False)
nonchinese_aids_top=Aids_unknown_sorted.head(10)
sns.barplot(x='Year',y='Ethnicity/Unknown',hue='Year',data=nonchinese_aids_top)
plt.title('Disease in Unknown tribes top 10 years')
plt.show()
```
